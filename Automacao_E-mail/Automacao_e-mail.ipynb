### Passo 4

Abrir o arquivo e pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python

# Calcular os indicadores
# abrir a base de dados 
# ver as informações da base de dados 
caminho = r'C:\Users\Tainá Bueno\Downloads\Vendas - Dez.xlsx'
tabela = pandas.read_excel(caminho)
display(tabela) #no jupyter a função display mostra a base de dados em um formato de tabela

# somar os indicadores 
faturamento = tabela["Valor Final"].sum()
qtde_produtos = tabela["Quantidade"].sum()

print(f'O valor total de faturamento no mês de Dezembro foi de R$ {faturamento:.2f}, e tivemos {qtde_produtos:,} produtos vendidos.')

```

### Passo 5
Enviar um e-mail pelo gmail

```python
# Enviar as informações por e-mail

# atalho para abrir uma nova aba
pyautogui.hotkey('ctrl', 't')

#entrar no e-mail
pyautogui.write("https://mail.google.com/")
pyautogui.press("enter")
time.sleep(5)

# clicar em novo e-mail
pyautogui.click(x=101, y=222)
time.sleep(2)

# para quem enviar o e-mail 
pyautogui.write('buenogoularttaina@gmai.com')
pyautogui.press("tab")
pyautogui.press("tab")

# qual o assunto 
pyperclip.copy('Relatório de Vendas')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press("tab")

# qual o corpo do email
texto = f"""    
Prezados,

Segue o relatório de vendas de hoje.

Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {qtde_produtos:,}

Qualquer dúvida estou à disposição.

Atenciosamente,
Tainá Goulart
"""
# escrever o texto no corpo do e-mail
pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')
time.sleep(3)
# enviar o e-mail
pyautogui.click(x=1321, y=1015)
```
